```python
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import cv2

# added - draw bounding box and label
CATEGORIES = [
    {
        'id': 1,
        'name': 'Leaf',
        'supercategory': '',
    }
]

image_directory = '/content/drive/MyDrive/ColabNotebooks/COCO_labelVis/SampleData_230522/'
annotation_file = '/content/drive/MyDrive/ColabNotebooks/COCO_labelVis/SampleData_230522/SampleData_COCO.json'

example_coco = COCO(annotation_file)

```

```python
categories = example_coco.loadCats(example_coco.getCatIds())
category_names = [category['name'] for category in categories]
print('Custom COCO categories: \n{}\n'.format(' '.join(category_names)))

```

```python
category_names = set([category['supercategory'] for category in categories])
print('Custom COCO supercategories: \n{}'.format(' '.join(category_names)))

category_ids = example_coco.getCatIds(catNms=['patternless'])
image_ids = example_coco.getImgIds(catIds=category_ids)
```

```python
for i in range(0,20):
  image_data = example_coco.loadImgs(image_ids)[i]
  # load and display instance annotations
  image = io.imread(image_directory + image_data['file_name'])
  img_name = image_data['file_name'].split('/')[2]

  plt.figure(figsize=(20,20))
  plt.subplot(1,2,1); plt.axis('off')
  plt.imshow(image)
  
  plt.subplot(1, 2, 2)      
  plt.imshow(image); plt.axis('off')
  pylab.rcParams['figure.figsize'] = (20.0, 20.0)

  # draw segmentation
  annotation_ids = example_coco.getAnnIds(imgIds=image_data['id'], catIds=category_ids, iscrowd=None)
  annotations = example_coco.loadAnns(annotation_ids)

  example_coco.showAnns(annotations, draw_bbox=False)
    
  plt.savefig('/content/drive/MyDrive/ColabNotebooks/COCO_labelVis/20230523TW_leaf_label_vis_result/2_'+f"{img_name.split('.')[0]}_annotated.jpg", bbox_inches="tight", pad_inches=0)
```
